```python
# Sort an Array of 0s, 1s, and 2s
This program sorts an array containing only 0s, 1s, and 2s using a one-pass algorithm (Dutch National Flag).
```

```python
class Solution:
    def sort012(self, arr):
        low = 0
        mid = 0
        high = len(arr) - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

        return arr


# Example run
arr = [0, 1, 2, 0, 1, 2]
print("Sorted array:", Solution().sort012(arr))
```

```python
# Binary Search (First Occurrence)
This program uses binary search to find the first occurrence of a given value in a sorted array.
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        result = -1

        while low <= high:
            mid = (low + high) // 2

            if arr[mid] == k:
                result = mid
                high = mid - 1
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1

        return result


# Example run
arr = [1, 1, 1, 1, 2]
k = 1
print("First occurrence index:", Solution().binarysearch(arr, k))
```
