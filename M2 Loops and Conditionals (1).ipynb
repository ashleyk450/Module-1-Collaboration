```python
## 4.1 Choose a number between 1 and 10
```

```python
secret = 6
guess = 4

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

```python
## 4.2 If/Else statements
```

```python
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

```python
## 6.1 For Loop - Print List Values
```

```python
numbers = [3, 2, 1, 0]

for num in numbers:
    print(num)
```

```python
## 6.2 Guess Me = 7
```

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("Too low")
    elif number == guess_me:
        print("Found it!")
        break
    else:
        print("Oops")
        break
    number += 1
```

```python
## 6.3 Guess Me = 5
```

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("Too low")
    elif number == guess_me:
        print("Found it!")
        break
    else:
        print("Oops")
        break
```
