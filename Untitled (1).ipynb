Used sqlalchemy to connect to books.db and selected titles in alphabetical order.

```python
from sqlalchemy import create_engine, text

engine = create_engine("sqlite:///books.db")
conn = engine.connect()

result = conn.execute(text("SELECT title FROM book ORDER BY title"))

for row in result:
    print(row[0])
```

```python
from sqlalchemy import create_engine, text

engine = create_engine("sqlite:///books.db")
conn = engine.connect()

# create table
conn.execute(text("""
CREATE TABLE book (
    title TEXT,
    author TEXT,
    year INTEGER
)
"""))

# insert data
conn.execute(text("""
INSERT INTO book VALUES
('Book A', 'Author A', 2000),
('Book B', 'Author B', 2005),
('Book C', 'Author C', 2010)
"""))

conn.commit()
```

```python
result = conn.execute(text("SELECT title FROM book ORDER BY title"))

for row in result:
    print(row[0])
```
